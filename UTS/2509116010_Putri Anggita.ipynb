###**Outliers Values**

```python
results = []

cols = depression.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = depression[col].quantile(0.25)
  q3 = depression[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = depression[(depression[col] < lower_bound) | (depression[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(depression))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_depression = pd.DataFrame(results)
results_depression.set_index('Kolom', inplace=True)
results_depression = results_depression.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_depression)
```

Dari analisis diatas, dapat disimpulkan bahwa terdapat 4 kolom yang memiliki *outliers values*. Nilai-nilai tersebut masih dalam bentuk proporsi yang apabila di kalkulasikan dalam bentuk persentase, maka nilainya adalah sebagai berikut:

1. Age sebesar 4,3%
2. Work Pressure 1,8%
3. CGPA 3,2%
4. Job Satisfaction 2.87%

Dilihat dari persentasenya, nilai-nilai tersebut dibawah 10% dan kemungkinan besar merupakan kesalahan atau anomali data yang harus ditangani dengan cara dihapus.

# **Eksplorasi Data (EDA)**

##**Comparison/Perbandingan**

* **Aktivitas**: Membandingkan rata-rata tingkat depresi pada setiap kategori tingkat kepuasan belajar (Study Satisfaction).

* **Tujuan**: Mengidentifikasi apakah terdapat perbedaan proporsi depresi pada masing-masing tingkat kepuasan belajar

* **Visualisasi**: Bar Chart (Grafik Batang)

```python
penjualan_market = depression.groupby('Study Satisfaction')['Depression'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))

sns.barplot(
    x=penjualan_market.index,
    y=penjualan_market.values,
    hue=penjualan_market.index,   # tambahkan hue
    palette='viridis',
    legend=False                  # matikan legend agar tidak dobel
)

plt.title('Study Satisfaction vs Depression')
plt.xlabel('Study Satisfaction')
plt.ylabel('Average Depression')
plt.xticks(rotation=90)

plt.show()
```

**Insight**

Berdasarkan grafik batang diatas, terlihat bahwa semakin tinggi tingkat kepuasan belajar, maka semakin rendah rata-rata depresinya. Hal ini menunjukkan adanya hubungan negatif antara kepuasan belajar dan depresi. Dari analisis tersebut, dapat disimpulkan bahwa pelajar yang merasa lebih puas terhadap proses belajarnya cenderung memiliki risiko depresi yang lebih rendah.

##**Relationship/Hubungan**

* **Aktivitas**: Menganalisis hubungan antara riwayat keluarga dengan penyakit mental dan depresi pada pelajar.

* **Tujuan**: Mengidentifikasi apakah pelajar yang memiliki riwayat keluarga dengan penyakit mental memiliki kecenderungan lebih tinggi untuk mengalami depresi.

* **Visualisasi**: Heatmap


```python

depression['Family History of Mental Illness'] = depression['Family History of Mental Illness'].map({'Yes': 1, 'No': 0})


corr_matrix = depression[['Family History of Mental Illness', 'Depression']].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Korelasi antara CGPA dan Potensi Depresi')
plt.show()
```

**Insight**

Nilai korelasi antara kedua variabel ini adalah 0.05 yang mana nilainya sangat rendah. Ini menunjukkan bahwa tidak ada hubungan yang signifikan antara memiliki keluarga dengan penyakit mental dan mengalami depresi di dalam data ini. Sehingga dapat disimpulkan bahwa faktor riwayat keluarga dengan penyakit mental bukan faktor penentu utama dalam perkembangan depresi pada pelajar.

**Work Pressure dan Job Satisfaction**

Pada kolom Work Pressure dan Job Satisfaction terdapat outliers dan perlu ditangani dengan cara imputasi, yaitu sebagai berikut:

```python
columns_to_impute = ["Work Pressure", "Job Satisfaction"]

for col in columns_to_impute:
    Q1 = depression[col].quantile(0.25)
    Q3 = depression[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    depression.loc[:, col] = depression[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = depression.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = depression[col].quantile(0.25)
  q3 = depression[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = depression[(depression[col] < lower_bound) | (depression[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(depression))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_depression = pd.DataFrame(results)
results_depression.set_index('Kolom', inplace=True)
results_depression = results_depression.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_depression)
```

Setelah dilakukan pengecekan kembali, penanganan outliers pada kolom Work Pressure dan Job Satisfaction berhasil dilakukan sementara untuk Age dan CGPA tidak dilakukan penanganan sesuai dengan alasan yang disebutkan.

##**Data Construction**

Data Construction adalah tahapan dimana kita membuat fitur baru untuk membantu menunjang kebutuhan analisis kedepannya. Dalam dataset ini, akan dibuat fitur tambahan, yaitu:

**1. Healthy Lifestyle**

Healthy Lifestyle adalah kolom yang dibuat untuk mengelompokkan individu berdasarkan keseimbangan pola hidup dengan mempertimbangkan tekanan akademik, stress finansial, kepuasan belajar, serta durasi belajar.

Tujuan pembuatan kolom ini adalah untuk mengidentifikasi apakah pola hidup individu ini sehat, menengah, atau berisiko depresi.

Kolom ini merupakan kombinasi dari 4 kolom, yaitu:


*   Academic Pressure
*   Financial Stress
*   Study Satisfaction
*   Work/Study Hours

Keempat kolom ini dikombinasikan agar dapat menggabungkan beberapa faktor menjadi satu kolom  yang dapat menunjukkan skala pola hidup pelajar karena terjadinya depresi tidak hanya dilihat dari satu faktor saja melainkan gabungan dari beberapa faktor.

Dari kolom ini, terdapat 3 kategori pola hidup pelajar

* Healthy

  Kategori ini terjadi ketika pelajar memiliki tekanan akademik dan finansial yang rendah, kepuasan belajar yang tinggi, dan durasi belajar yang normal.

* Moderate

  Kategori ini terjadi ketika pelajar memiliki tekanan akademik dan finansial yang cukup tinggi, tetapi seimbang dengan kepuasan dan durasi belajar yang masih normal.

* Risky

  Kategori ini terjadi ketika pelajar memiliki tekanan akademik dan finansial yang tinggi diikuti dengan kepuasan dan durasi belajar yang juga tinggi sehingga lebih berisiko besar terkena depresi.



```python
cols = ['Academic Pressure', 'Financial Stress',
        'Study Satisfaction', 'Work/Study Hours']

for col in cols:
    depression[col] = pd.to_numeric(depression[col], errors='coerce')

def lifestyle_category(row):
    total_stress = row['Academic Pressure'] + row['Financial Stress']

    if (total_stress <= 4) and (row['Study Satisfaction'] >= 4) and (row['Work/Study Hours'] <= 8):
        return "Healthy"

    elif (total_stress >= 7) or (row['Study Satisfaction'] <= 2) or (row['Work/Study Hours'] > 8):
        return "Risky"

    else:
        return "Moderate"

depression['Healthy_Lifestyle'] = depression.apply(lifestyle_category, axis=1)

print(depression[['Academic Pressure', 'Financial Stress',
                  'Study Satisfaction', 'Work/Study Hours',
                  'Healthy_Lifestyle']].head())
```

Sebagian kecil pelajar dengan jam belajar rendah juga tetap mengalami depresi sehingga dapat disimpulkan bahwa durasi belajar bukan satu-satunya faktor yang menyebabkan tingginya tekanan yang dialami oleh pelajar.

#**Heatmap**

* **Aktivitas**: Melihat korelasi antara berbagai faktor penyebab depresi pada siswa.

* **Tujuan**: Untuk menganalisis sekuat apa faktor-faktor yang ada dalam meningkatkan risiko depresi.

* **Visualisasi**: Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=depression.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Korelasi Faktor Depresi Pelajar')
plt.show()
```

**Insight**

Dari keseluruhan variabel numerik diatas, insight yang didapat yaitu:



1. Academic Pressure dan Depression berada di angka 0.47 dan dikategorikan dalam korelasi positif sedang. Hal ini berarti semakin tinggi tekanan akademik pelajar, maka potensi terkena depresi semakin meningkat.

2. Financial Stress dan Depression berada di angka 0.36 dan termasuk kategori korelasi positif lemah. Karena korelasi sifatnya positif, maka semakin tinggi tekanan finansial, maka semakin tinggi pula potensi terkena depresi nya.

3. Work/Study Hours dan Depression berada di angka 0.21 dan dikategorikan sebagai korelasi positif lemah. Hal ini berarti semakin tinggi durasi waktu belajar, maka semakin tinggi pula potensi terkena depresi.

4. Age dan Depression berada di angka -0.23 dan dikategorikan sebagai korelasi negatif lemah. Karena sifat korelasinya negatif, maka apabila salah satu variabel meningkat, maka variabel lainnya akan menurun. Dalam hal ini, semakin muda usia seorang pelajar, maka semakin rentan mengalami depresi dibandingkan yang lebih tua.

5. Study Satisfaction dan Depression berada di angka -0.17 dan dikategorikan sebagai korelasi negatif sangat lemah. Artinya, apabila kepuasan belajar meningkat, maka potensi terkena depresi akan cenderung menurun.

6. CGPA dan Depression berada di angka 0.02. Angka ini menunjukkan korelasi sangat lemah dan hampir tidak ada hubungannya. Dari sini dapat disimpulkan bahwa IPK tidak mempengaruhi tingkat depresi pelajar.

7. Family History Of Mental Illness dan Depression berada di angka 0.05. Meskipun awalnya dicurigai bahwa penyakit depresi ini diturunkan secara genetik. Nyatanya, dalam data ini riyawat penyakit mental dalam keluarga dengan potensi depresi pelajar hampir tidak ada hubungannya.
